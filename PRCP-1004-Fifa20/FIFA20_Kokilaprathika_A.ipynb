Task 2 – Clustering Players Based on Football Skills


```python
# Feature Selection
skill_columns=['attacking_finishing','attacking_short_passing', 'attacking_volleys','skill_dribbling','skill_curve','skill_fk_accuracy','skill_long_passing','skill_ball_control','movement_acceleration','movement_sprint_speed','movement_agility','movement_reactions','movement_balance','power_shot_power','power_jumping','power_stamina','power_strength','power_long_shots','mentality_aggression','mentality_positioning','mentality_vision','mentality_composure','attacking_crossing','attacking_finishing','attacking_heading_accuracy']
skill_columns=[col for col in skill_columns if col in df.columns]
skills=df[skill_columns]
skills.head()
```

```python
skills=df[skill_columns]
```

```python
skills.info()
skills.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler=StandardScaler()
skills_scaled=scaler.fit_transform(skills)
```

KMeans Cluster

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
inertia=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(skills_scaled)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),inertia,marker='o')
plt.title("Elbow Method _ Optimal K")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,random_state=42)
kmeans.fit(skills_scaled)
```

```python
cluster_centers=scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
cluster_centers_df=pd.DataFrame(cluster_centers,columns=skill_columns)
cluster_centers_df
```

```python
df['cluster']=kmeans.labels_
df[['short_name','cluster']].head()
```

```python
df['cluster'].value_counts()
```

```python
cluster_profile=df.groupby('cluster')[skill_columns].mean()
cluster_profile
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=df['dribbling'],y=df['attacking_finishing'],hue=df['cluster'])
plt.title("players clusters based on Dribbling & Finishing")
plt.xlabel("'Dribbiling")
plt.ylabel("Finishing")
plt.show()
```

```python
cluster_names={0:"Winger Type",1:"Finisher/Stiker",2:"Playmaker",3:"Physical Forward"}
df['cluster_name']=df['cluster'].map(cluster_names)
df[['short_name','cluster','cluster_name']].head()
```
